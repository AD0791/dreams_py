# vbg scenarios

```python
vbg_cc = read_excel('vbg_fy21_CC.xlsx', parse_dates=True)
vbg_cc.rename(columns={'dreams_code':'code'},inplace=True)
vbg_cc.code.count()
```

```python
vbg_ci = read_excel('vbg_fy21_CI.xlsx', parse_dates=True)
vbg_ci.rename(columns={'dreams_code':'code'},inplace=True)
vbg_ci.code.count()
```

```python
vbg_nc = read_excel('vbg_fy21_NC.xlsx', parse_dates=True)
vbg_nc.rename(columns={'dreams_code':'code'},inplace=True)
vbg_nc.code.count()
```

```python
code_vbgcc = vbg_cc[['code']]
code_vbgci = vbg_ci[['code']]
code_vbgnc = vbg_nc[['code']]
```

```python
vbgcc = code_vbgcc.merge(base,how='left',on='code').fillna('NULL')
vbgci = code_vbgci.merge(base,how='left',on='code').fillna('NULL') 
vbgnc = code_vbgnc.merge(base,how='left',on='code').fillna('NULL') 
```

```python


curr_vbgcc = vbgcc[vbgcc.age_range!='NULL'].pivot_table(index='age_range',columns="dreams_curriculum",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
curr_vbgcc

cond_vbgcc = vbgcc[vbgcc.age_range!='NULL'].pivot_table(index='age_range',columns="new_condoms",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
cond_vbgcc
cond_vbgcc_liste = vbgcc[(vbgcc.age_range!='NULL')&(vbgcc.new_condoms=='no')]


hts_vbgcc = vbgcc[vbgcc.age_range!='NULL'].pivot_table(index='age_range',columns="hts",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
hts_vbgcc
hts_vbgcc_liste = vbgcc[(vbgcc.age_range!='NULL')&(vbgcc.hts=='errata')]

vbgcc[vbgcc.age_range!='NULL'].pivot_table(index='age_range',columns="sens_prep",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
vbgcc[vbgcc.age_range!='NULL'].pivot_table(index='age_range',columns="acc_prep",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
vbgcc[vbgcc.age_range!='NULL'].pivot_table(index='age_range',columns="ref_prep",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
iprep_vbgcc = vbgcc[vbgcc.age_range!='NULL'].pivot_table(index='age_range',columns="init_prep",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
iprep_vbgcc
```

```python


curr_vbgci = vbgci.pivot_table(index='age_range',columns="dreams_curriculum",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
curr_vbgci

cond_vbgci = vbgci.pivot_table(index='age_range',columns="new_condoms",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
cond_vbgci
cond_vbgci_liste = vbgci[(vbgci.new_condoms=='no')]


hts_vbgci = vbgci.pivot_table(index='age_range',columns="hts",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
hts_vbgci
hts_vbgci_liste = vbgci[(vbgci.hts=='errata')]

vbgci.pivot_table(index='age_range',columns="sens_prep",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
vbgci.pivot_table(index='age_range',columns="acc_prep",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
vbgci.pivot_table(index='age_range',columns="ref_prep",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
iprep_vbgci = vbgci.pivot_table(index='age_range',columns="init_prep",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
iprep_vbgci
```

```python


curr_vbgnc = vbgnc.pivot_table(index='age_range',columns="dreams_curriculum",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
curr_vbgnc

cond_vbgnc = vbgnc.pivot_table(index='age_range',columns="new_condoms",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
cond_vbgnc
cond_vbgnc_liste = vbgnc[(vbgnc.new_condoms=='no')]


hts_vbgnc = vbgnc.pivot_table(index='age_range',columns="hts",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
hts_vbgnc
hts_vbgnc_liste = vbgnc[(vbgnc.hts=='errata')]

vbgnc.pivot_table(index='age_range',columns="sens_prep",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
vbgnc.pivot_table(index='age_range',columns="acc_prep",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
vbgnc.pivot_table(index='age_range',columns="ref_prep",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
iprep_vbgnc = vbgnc.pivot_table(index='age_range',columns="init_prep",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
iprep_vbgnc
```

# Depistage

```python

depistage_base = read_excel('rapportage_liste_filles_hts_restant_24_08_2021.xlsx',sheet_name='liste_des_filles_global',parse_dates=True)
depistage_base.code.count()
depistage = depistage_base[['code']].merge(base,how='left',on='code').fillna('NULL')
depistage.code.count()
```

```python
missing_depistage = depistage[depistage.newage_range=="NULL"]
missing_depistage.code.count()
#missing_depistage = missing_depistage[['code']].merge(depistage,how='left',on='code')
missing_depistage = depistage_base[depistage_base.code.isin(missing_depistage.code)]
missing_depistage.code.count()
```

```python



curr_depistage = depistage[depistage.newage_range!="NULL"].pivot_table(index='newage_range',columns="curriculum",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
curr_depistage
curr_depistage_I_liste = depistage[(depistage.newage_range!="NULL")&(depistage.curriculum=='curriculum incomplet')]
curr_depistage_NR_liste = depistage[(depistage.newage_range!="NULL")&(depistage.curriculum=='non-recu')]



cond_depistage = depistage[depistage.newage_range!="NULL"].pivot_table(index='newage_range',columns="new_condoms",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
cond_depistage
cond_depistage_liste = depistage[(depistage.newage_range!="NULL")&(depistage.new_condoms=='no')]


depistage[depistage.newage_range!="NULL"].pivot_table(index='newage_range',columns="sens_prep",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
depistage[depistage.newage_range!="NULL"].pivot_table(index='newage_range',columns="acc_prep",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
depistage[depistage.newage_range!="NULL"].pivot_table(index='newage_range',columns="ref_prep",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
iprep_depistage = depistage[depistage.newage_range!="NULL"].pivot_table(index='newage_range',columns="init_prep",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
iprep_depistage
iprep_depistage_liste = depistage[(depistage.newage_range!="NULL")&(depistage.init_prep=='errata')]
```

```python



curr_missing_depistage = missing_depistage[missing_depistage.newage_range!="NULL"].pivot_table(index='newage_range',columns="curriculum",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
curr_missing_depistage
curr_missing_depistage_I_liste = missing_depistage[(missing_depistage.newage_range!="NULL")&(missing_depistage.curriculum=='curriculum incomplet')]
curr_missing_depistage_NR_liste = missing_depistage[(missing_depistage.newage_range!="NULL")&(missing_depistage.curriculum=='non-recu')]



cond_missing_depistage = missing_depistage[missing_depistage.newage_range!="NULL"].pivot_table(index='newage_range',columns="condoms",
                        values='code',
                        aggfunc='count',
                        fill_value=0,
                        margins=True,
                        margins_name="Grand Total"                             
)
cond_missing_depistage
cond_missing_depistage_liste = missing_depistage[(missing_depistage.newage_range!="NULL")&(missing_depistage.condoms=='errata')]

```
